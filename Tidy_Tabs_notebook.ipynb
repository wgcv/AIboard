````python
import requests

# Your GitHub API token
token = '<TOKEN>'

# Headers for authentication
headers = {
    'Authorization': f'token {token}'
}

# Base URL for GitHub API
base_url = 'https://api.github.com'

# Function to get user URLs
def get_user_urls(num_users=100):
    user_urls = []
    page = 1
    per_page = 50  # GitHub API allows up to 100 results per page

    while len(user_urls) < num_users:
        response = requests.get(f'{base_url}/users', headers=headers, params={'per_page': per_page, 'page': page})
        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        users = response.json()
        if not users:
            break

        for user in users:
            user_urls.append({"url": user['html_url'], "summary": user['login'] +" GitHub Profile"})
            if len(user_urls) >= num_users:
                break

        page += 1

    return user_urls

# Collect 100 user URLs
user_urls = get_user_urls(10000)

github_profile = []
# Print the URLs
for url in user_urls:
    github_profile.append(url)
````

````python
import requests
import random
import time

# Your New York Times API key
api_key = '<token>'

# Base URL for the Article Search API
base_url = 'https://api.nytimes.com/svc/search/v2/articlesearch.json'

# Function to get random article URLs
def get_random_article_urls(num_articles=100, page = 0):
    article_urls = []

    while len(article_urls) < num_articles:
        # Randomize the page parameter to get articles from different pages
        params = {
            'api-key': api_key,
            'page': page,
            'fq': 'document_type:("article")',
            'page_size': 10,
            'sort': 'newest'  # Sort by newest articles

        }

        response = requests.get(base_url, params=params)
        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        articles = response.json().get('response', {}).get('docs', [])
        if not articles:
            break

        for article in articles:
            article_urls.append(article['web_url'])
            if len(article_urls) >= num_articles:
                break

        # To avoid hitting the rate limit, sleep for a bit
        time.sleep(2)
        page +=1
        print(page)
        if(page % 5 == 0):
          time.sleep(60)
    return article_urls

# Collect 100 random article URLs
article_urls1 = get_random_article_urls(200, 93)

````

### Function to load the html

````python
import requests
def get_content(url):
  # Make a request to Prerender.io
  response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36'})

  # Process the response
  html_content = response.text

  # Parse the HTML content
  soup = BeautifulSoup(html_content, 'html.parser')

  # Extract the title
  title = soup.title.string if soup.title else ''

  # Extract meta description
  meta_description = soup.find('meta', attrs={'name': 'description'})
  description = meta_description['content'] if meta_description else ''

  # Extract headings
  headings = [h.get_text() for h in soup.find_all(['h1', 'h2', 'h3'])]

  # Extract main paragraphs
  paragraphs = [p.get_text() for p in soup.find_all('p')]
  headings = ' '.join(headings)
  paragraphs = ' '.join(paragraphs)
  headings = headings.replace("\n", "").replace("\t", "").replace(",", ";")
  paragraphs = headings.replace("\n", "").replace("\t", "").replace(",", ";")
  description = description.replace(",", ";")
  title = title.replace(",", ";")
  text = "[title] "+title + "\n [description]" + description
  return {"url": url, "title":title, "description": description,  "paragraphs": paragraphs, "headings":headings, "text": text , "summary": ""}
````

````python
import evaluate

rouge = evaluate.load("rouge")
````

````python
import numpy as np


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)

    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
````

````python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
# Add the new tokens
model.resize_token_embeddings(len(tokenizer))
````

````python
training_args = Seq2SeqTrainingArguments(
    output_dir="./tidy-tab-model-t5-small",
    hub_model_id='wgcv/tidy-tab-model-t5-small',
    evaluation_strategy="steps",
    save_strategy="steps",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=5,
    num_train_epochs=32,
    predict_with_generate=True,
    fp16=False,
    push_to_hub=True,
    load_best_model_at_end=True,
    warmup_steps = 1000
    )

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
      callbacks=[early_stopping_callback]

)

trainer.train()
````
